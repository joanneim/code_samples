## Question 1

*(a) Return of stock X has mean 20% and volatility 50%. Market return has mean 8%
and volatility 20%. The correlation between stock X and market is 50%. If we
run regression of stock X return on market return (with intercept), what is the
regression coefficient on market return (i.e. beta)?*

Let $R_X$ denote the return on stock $X$ and $R_M$ denote the return on the market. The regression coefficient in this univariate regression can be expressed as

\begin{eqnarray}
\beta & = & \frac{\mathrm{cov} \left(R_X, R_M \right)}{\mathrm{var}\left(R_M\right)} \\
 & = & \frac{\rho_{XM} \sigma_{X} \sigma_{M}}{\mathrm{var}\left(R_M\right)} \\
 & = & \rho_{XM} \frac{\sigma_{X}}{\sigma_M} \\
 & = & 1.25
\end{eqnarray}

where the first equality holds by the fact that we are running a univariate regression and the second equality holds by definition.

*(b) Would you use daily, weekly, or monthly returns to estimate the market beta of a stock? Please explain.*

There are multiple factors to consider when deciding on the choice of return interval.

Using returns at a daily frequency provides more observations, and therefore, allows more precise estimation of parameters. This is particularly relevant if returns are approximately IID. At the same time, returns at a daily frequency may contain more noise, in particular, if the stock trades infrequently.

Naturally, it is not possible to capture high-frequency dynamics with returns at a low-frequency. If market dynamics are changing rapidly, it might be better to use daily returns to model high-frequency dynamics. Meanwhile, if dynamics are changing at a low-frequency, using low-frequency returns might be more appropriate.

*(c) You want to predict Walmart’s sales using retail traffic activities. You just learned that one can count the number of cars in Walmart’s parking lots using satellite data. Clearly, the car counts are a very noisy measure of the actual level of retail
traffic. If you run a linear regression of sales on car counts, how would the noise affect the results?*

We can formalize this setup as follows. Consider the following linear regression model:

$$y_t = \alpha + \beta X_t + \epsilon_t, \quad \mathbb{E} \left[\epsilon_t \mid X_t\right] = 0$$

where $y_t$ denote Walmart's sales at time $t$, $X_t$ denote retail traffic activities, and $\epsilon_t$ is an error term satisfying $\mathbb{E} \left[\epsilon_t^2 \right] = \sigma_{\epsilon}^2$.

Under these assumptions, the population coefficient in the regression above satisfies

$$\beta = \frac{\sigma_{X, y}}{\sigma_X^2}$$

where $\sigma_{X, y} = \mathrm{Cov} \left(X_t, y_t\right)$ and $\sigma_X^2 = \mathrm{Var} \left(X_t\right)$.

Further, assume that $\widetilde{X}_t = X_t + \nu_t$ where $\widetilde{X}_t$ denotes car count and $\nu_t$ is an error term capturing the noise in the measure of the actual level of retail traffic that satisfies $\mathbb{E} \left[\nu_t\right] = 0$, $\mathbb{E} \left[\nu_t^2\right] = \sigma_{\nu}^2$, and $\nu_t \perp \epsilon_t$.

Suppose that we run a regression of $y_t$ on $\widetilde{X}_t$ and a constant. The population regression coefficient satisfies

\begin{eqnarray}\widetilde{\beta} & = & \frac{\mathrm{Cov} \left(X_t + \nu_t, y_t \right)}{\mathrm{Var} \left(X_t \right)} \\
& = & \frac{\sigma_{X, y}}{\sigma_X^2 + \sigma_{\nu}^2} \\
& = & \frac{\sigma_X^2}{\sigma_X^2 + \sigma_{\nu}^2} \beta
\end{eqnarray}

Note that this implies $\left|\widetilde{\beta} \right| < \left| \beta \right|$.

Under our assumptions, OLS provides an unbiased estimate of $\widetilde{\beta}$. Therefore, this implies that it provides an estimate of $\beta$ that is biased towards zero.



## Question 2

*(a) Consider a simple linear regression $y_i = \beta_0 + \beta_1 x_i +ε_i$. Show that the $R^2$ statistic of this regression equals to the squared value of the correlation between $x$ and $y$,*

$$R^2 = \widehat{\mathrm{Corr}} \left(x ,y\right) ^ 2$$

*where*

$$\widehat{\mathrm{Corr}}(x, y) = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}
{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^{n} (y_i - \bar{y})^2}}$$

In a simple regression, we have

\begin{eqnarray}
  \widehat{\beta}_0 & = & \bar{y} - \widehat{\beta}_1 \bar{x} \\
  \widehat{\beta}_1 & = & \frac{\sum_{i=1}^n \left(x_i - \bar{x}\right) \left(y_i - \bar{y}\right)}{\sum_{i=1}^n \left(x_i - \bar{x}\right)^2}
\end{eqnarray}

We can compute $R^2$ as

\begin{eqnarray}
  R^2 & = & 1 - \frac{\sum_{i=1}^n \left(y_i - \widehat{y}_i\right)^2}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \\
  & = & 1 - \frac{\sum_{i=1}^n \left(y_i - \bar{y} - \widehat{\beta}_1 \left(x_i - \bar{x} \right)\right)^2}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \\
  & = & 1 - \frac{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2 - 2 \left(y_i - \bar{y}\right) \widehat{\beta}_1 \left(x_i - \bar{x} \right)  + \left(\widehat{\beta}_1 \left(x_i - \bar{x} \right)\right)^2}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \\
  & = & \frac{1}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \left(\sum_{i=1}^n 2 \left(y_i - \bar{y}\right) \widehat{\beta}_1 \left(x_i - \bar{x} \right) - \left(\widehat{\beta}_1 \left(x_i - \bar{x} \right)\right)^2\right) \\
  & = & \frac{1}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \left(2 \frac{\left(\sum_{i=1}^n  \left(y_i - \bar{y}\right)  \left(x_i - \bar{x} \right)\right)^2}{\sum_{i=1}^n \left(x_i - \bar{x}\right)^2} - \frac{\left(\sum_{i=1}^n  \left(y_i - \bar{y}\right)  \left(x_i - \bar{x} \right)\right)^2}{\sum_{i=1}^n \left(x_i - \bar{x}\right)^2}\right) \\
  & = & \frac{\left(\sum_{i=1}^n  \left(y_i - \bar{y}\right)  \left(x_i - \bar{x} \right)\right)^2}{\left(\sum_{i=1}^n \left(x_i - \bar{x}\right)^2\right) \left(\sum_{i=1}^n \left(y_i - \bar{y}\right)^2\right)} \\
  & = & \widehat{\mathrm{Corr}} \left(x, y\right) ^ 2
\end{eqnarray}

where the first equality follows from the definition of $R^2$, the second equality uses the expression for $\widehat{\beta}_0$, the fifth equality uses the expression for $\widehat{\beta}_1$, and the seventh equality follows from the defintion of $\widehat{\mathrm{Corr}}$.

*(b) Now let's extend the result to multiple linear regression. Show that*

$$
R^2 = \widehat{\mathrm{Corr}}\left(\widehat{y}, y\right)^2,
$$

*where $\widehat{y}_i = x'_i \widehat{\beta}$ is the predicted value for $y_i$.*

Following steps similar to (a), we have

\begin{eqnarray}
  R^2 & = & 1 - \frac{\sum_{i=1}^n \left(y_i - \widehat{y}_i\right)^2}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \\
  & = & 1 - \frac{\sum_{i=1}^n \left(y_i - \bar{y} + \bar{y} - \widehat{y}_i\right)^2}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \\
  & = & \frac{\sum_{i=1}^n 2 \left(y_i - \bar{y}\right) \left(\widehat{y}_i - \bar{y}\right) - \left(\widehat{y}_i - \bar{y}\right)^2}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2} \\
  & = & \frac{\sum_{i=1}^n \left(\widehat{y}_i - \bar{y}\right)^2}{\sum_{i=1}^n  \left(\widehat{y}_i - \bar{y}\right)^2} \frac{\sum_{i=1}^n 2 \left(y_i - \bar{y}\right) \left(\widehat{y}_i - \bar{y}\right) - \left(\widehat{y}_i - \bar{y}\right)^2}{\sum_{i=1}^n \left(y_i - \bar{y}\right)^2}
\end{eqnarray}

Observe that the desired result holds if $\sum_{i=1}^n  \left(\widehat{y}_i - \bar{y}\right)^2 = \sum_{i=1}^n \left(y_i - \bar{y} \right) \left(\widehat{y}_i - \bar{y}\right)$. We can prove this equality as follows

\begin{eqnarray}
  \sum_{i=1}^n \left(y_i - \bar{y} \right) \left(\widehat{y}_i - \bar{y}\right) & = & \sum_{i=1}^n \left(\widehat{y}_i + \widehat{\epsilon}_i - \bar{y} \right) \left(\widehat{y}_i - \bar{y}\right) \\
  & = & \sum_{i=1}^n \left(\widehat{y}_i - \bar{y} \right) ^ 2 + \sum_{i=1}^n \widehat{\epsilon}_i \widehat{y}_i + \sum_{i=1}^n \widehat{\epsilon}_i \bar{y} \\
  & = & \sum_{i=1}^n  \left(\widehat{y}_i - \bar{y}\right)^2
\end{eqnarray}

where the first equality follows by definition, and the third equality follows from the first-order condition for the OLS estimator (assuming a constant is included in $X$).


## Question 3

*(a) Prove that the following claim from the lecture notes is true: "When irrelevant variables are included, the LS estimator is still unbiased." Specifically, suppose the true DGP is:*

$$
Y = X_1\beta_1 + \epsilon,
$$

*but we are instead estimating:*

$$
Y = X_1\beta_1 + X_2\beta_2 + \epsilon.
$$

*Show that $\mathbb{E}\left[\widehat{\beta}_{1} \mid X \right] = \beta_1$.*

Let $X = \begin{bmatrix} X_1 & X_2 \end{bmatrix}$. We can express the least squares estimator as

\begin{eqnarray}
\widehat{\beta} & = & \begin{bmatrix} \widehat{\beta}_1 \\ \widehat{\beta}_2 \end{bmatrix} \\
& = & \left(X'X\right)^{-1} X' Y
\end{eqnarray}

Therefore,

\begin{eqnarray}
  \begin{bmatrix} \mathbb{E} \left[\widehat{\beta}_1 \mid X \right] \\ \mathbb{E} \left[\widehat{\beta}_2 \mid X \right] \end{bmatrix} & = & \mathbb{E} \left[\widehat{\beta} \mid X \right] \\ & = &   \left(X'X\right)^{-1} X' \mathbb{E} \left[Y \mid X \right] \\
  & = & \left(X'X\right)^{-1} X' \mathbb{E} \left(X \beta + \mathbb{E} \left[ \epsilon \mid X \right]\right) \\
  & = & \beta \\
  & = & \begin{bmatrix} \beta_1 \\ \beta_2 \end{bmatrix}
\end{eqnarray}

as required.

*(b) Explain why the inclusion of irrelevant feature $ X_2 $ could reduce the precision of LS estimate $\widehat{\beta}_1$.*


In the lecture notes, we have shown that $\mathrm{Var} \left(\widehat{\beta} \mid X \right) = \sigma^{2} \left(X' X\right)^{-1}$.

When running a univariate regression on $X_1$, this implies that the precision of $\widehat{\beta}_1$ conditional on $X_1$ is $\sigma^{-2} X_1' X_1$.

In a bivariate regression on $X_1$ and $X_2$, the conditional variance of the least squares estimator is

\begin{eqnarray}
  \mathrm{Var} \left(\widehat{\beta} \mid X \right) & = & \sigma^{2} \left(X' X\right)^{-1} \\
  & = & \sigma^{2} \frac{1}{\left(X_1' X_1\right) \left(X_2' X_2\right) - \left(X_1' X_2\right) ^ 2} \begin{bmatrix} X_2' X_2 & -X_2'X_1 \\
  -X_1'X_2 & X_1' X_1 \end{bmatrix}
\end{eqnarray}

where the first equality follows from the result in the lecture notes and the assumption that $X_2$ is an irrelevant feature.

Therefore, the precision of $\widehat{\beta}_1$ is $\sigma^{-2} \left(X_1'X_1 - \frac{\left(X_1' X_2\right)^2}{X_2' X_2}\right)$.

If $X_1' X_2 > 0$, then the precision of $\widehat{\beta}_1$ is lower in the case of the bivariate regression.

*(a) From Ken French’s website (https://goo.gl/TrqEYX), download (i) monthly market
excess returns and one-month Treasury bill rates, and (ii) monthly returns of
the 30 industry portfolios. We will define the “full sample” as the period 01/1971

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.regression.rolling import RollingOLS
from io import BytesIO
import zipfile
from urllib.request import urlopen
```

```python
url_mkt = urlopen("http://mba.tuck.dartmouth.edu/pages/faculty/ken.french/ftp/F-F_Research_Data_Factors_CSV.zip")
zfile = zipfile.ZipFile(BytesIO(url_mkt.read()))
df_ff = (pd.read_csv(zfile.open('F-F_Research_Data_Factors.CSV'),
                     header = 0,
                     names = ['date','exmkt','smb','hml','rf'],
                     skiprows=3,
                     nrows=1158,
                     index_col=0,
                     parse_dates=True,
                     date_parser=lambda x: pd.to_datetime(x, format='%Y%m'))
          .apply(lambda x: x / 100)
          .loc['1971-01-01':'2022-12-01']
          )


url_ind = urlopen("http://mba.tuck.dartmouth.edu/pages/faculty/ken.french/ftp/30_Industry_Portfolios_CSV.zip")
zfile = zipfile.ZipFile(BytesIO(url_ind.read()))
ind = (pd.read_csv(zfile.open('30_Industry_Portfolios.CSV'),
                  header=0,
                  skiprows=11,
                  nrows=1159,
                  index_col=0,
                  parse_dates=True,
                  date_parser=lambda x: pd.to_datetime(x, format='%Y%m'))
       .apply(lambda x: x / 100 - df_ff.rf)
       .loc['1971-01-01':'2022-12-01']
       )
```

*(b) Use the full sample, fit the market model to each of the 30 industry portfolios. Plot the data and fitted regression line (as in the lecture notes) for the top 5 and bottom 5 industries ranked by their market betas.
```python
def get_coefs(portfolio):
  result = sm.OLS(portfolio, sm.add_constant(df_ff.exmkt)).fit()
  coefs = pd.Series({"alpha": result.params[0], "beta": result.params[1]})
  return coefs

coefs = ind.apply(get_coefs).T
coefs_sorted = coefs.sort_values(by="beta")
```

*(c) We would like to form a "cyclical sector index" as an equal-weight portfolio of the top 5 industries by market beta, and a "defensive sector index" as an equal-weight portfolio of the bottom 5 industries. Unlike in Part (b), we want to sort the industries dynamically so as to capture the potential changes in cyclicality for a given sector over time. So here is what we need to do:* \\

*i. Starting in 01/1981, at the beginning of each month, use data from the previous 5 years (this is called "rolling-window") to estimate the industry betas
and then update the two indices according to the new market beta ranking. (If the components of the top and bottom industries change, you would be rebalancing the portfolio by selling the existing positions on some industries and buying stocks of new industries.)*

*ii. Report the average excess return, market beta, alpha, Sharpe ratio, information ratio (using market portfolio as benchmark), and maximum drawdown for the two portfolios. Comment on your findings.*

```python
def compute_drawdown(returns):
    """
    Calculate the drawdown of a series of returns.

    Parameters:
    - returns (array-like): The series of returns for which the drawdown is to
                            be computed.

    Returns:
    - array: The drawdown series which is the ratio of cumulative returns to
             the running maximum.
    """
    cum_returns = np.multiply.accumulate(1 + returns)
    run_max = np.maximum.accumulate(cum_returns)
    drawdown = cum_returns / run_max

    return drawdown


def get_coefs_rolling(portfolio):
  estimator = sm.regression.rolling.RollingOLS
  result = estimator(100 * portfolio,
                     sm.add_constant(100 * df_ff.exmkt),
                     window=60).fit()

  result.params.columns = (('alpha', portfolio.name), ('beta', portfolio.name))

  return result.params


def print_stats(index, weights):
  avg_excess_return = index.mean()
  avg_alpha = np.sum(rolling_alphas * weights, axis=1).mean()
  avg_beta = np.sum(rolling_betas * weights, axis=1).mean()
  sharpe_ratio = index.mean() / index.std()
  tracking_error = index - df_ff.exmkt.loc[index.index]
  info_ratio = tracking_error.mean() / tracking_error.std()
  drawdown = compute_drawdown(index)

  stats = {
        'Average Excess return': avg_excess_return,
        'Average Alpha': avg_alpha,
        'Average Beta': avg_beta,
        'Sharpe Ratio': sharpe_ratio,
        'Information Ratio': info_ratio,
        'Max Drawdown': (drawdown.min() - 1) * 100
  }

  print(f"Index: {index.name}")

  for key, value in stats.items():
      print(f"{key}: {value}")

  print("")


rolling_coefs = (pd.concat([get_coefs_rolling(p) for _, p in ind.items()],
                            axis=1)
                 .dropna()
                 .shift()
                 .loc["1981":]
                 )
rolling_alphas = rolling_coefs['alpha']
rolling_betas = rolling_coefs['beta']
ranks = rolling_betas.rank(axis=1)
defensive_weights = ranks.apply(lambda x: 1 / 5 * (x <= 5))
cyclical_weights = ranks.apply(lambda x: 1 / 5 * (x >= 26))
defensive_index = np.sum(ind.loc[defensive_weights.index] * defensive_weights,
                         axis=1)
cyclical_index = np.sum(ind.loc[cyclical_weights.index] * cyclical_weights,
                         axis=1)
defensive_index.name = 'defensive'
cyclical_index.name = 'cyclical'
```

```python
print_stats(defensive_index, defensive_weights)
print_stats(cyclical_index, cyclical_weights)
```

*(d) Recall the market timing strategy we built in Q4 of Pset1. Assume that we have a mandate to be fully invested in stocks (no cash holding) at all times. Use the model in Q4(e), design a portfolio that is invested in the two sector indices constructed above and rebalanced at the end of each month, where the weights on the cyclical and defensive indices depend on the expected market excess return in each month. Compare your portfolio’s performance with the one from Pset1 for the period of 01/2014 to 12/2022.*

```python
def compute_nav(mu_t, sigma2_t, alpha):
  omega = mu_t / (alpha * sigma2_t)
  returns = 1 + df_ff.rf.loc["2014":] + omega * df_ff.exmkt.loc["2014":]
  nav = np.multiply.accumulate(returns)

  return nav

result = sm.tsa.AutoReg(df_ff.exmkt.loc[:"2013"], 1).fit()
mu_hat_t_ar = result.apply(df_ff.exmkt).fittedvalues.loc["2014":]
omega_alpha_1 = mu_hat_t_ar / result.sigma2
omega_alpha_10 = omega_alpha_1 / 10

returns_alpha_1 = 1 + omega_alpha_1 * cyclical_index + (1 - omega_alpha_1) * defensive_index
returns_alpha_10 = 1 + omega_alpha_10 * cyclical_index + (1 - omega_alpha_10) * defensive_index

nav_alpha_1 = np.multiply.accumulate(returns_alpha_1.dropna())
nav_alpha_10 = np.multiply.accumulate(returns_alpha_10.dropna())


alpha = 1
nav_ar_alpha_1 = compute_nav(mu_hat_t_ar, result.sigma2, alpha)

alpha = 10
nav_ar_alpha_10 = compute_nav(mu_hat_t_ar, result.sigma2, alpha)
```

```python
nav_alpha_1.plot(color='blue', label="Cyclical Strategy")
nav_ar_alpha_1.plot(color='red', label="AR Model")
plt.legend()
plt.title(r"$\alpha = 1$");
```

```python
nav_alpha_10.plot(color='blue', label="Cyclical Strategy")
nav_ar_alpha_10.plot(color='red', label="AR Model")
plt.legend()
plt.title(r"$\alpha = 10$");
```

The cyclical strategy performs better than the AR model when $\alpha=10$.
